```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# File to Load (Remember to Change These)
mouse_drug_data_to_load = "Resources/mouse_drug_data.csv"
clinical_trial_data_to_load = "Resources/clinicaltrial_data.csv"
```

```python
# Read the Mouse/Drug Data and the Clinical Trial Data
tumor_data = pd.read_csv(mouse_drug_data_to_load)
drug_data = pd.read_csv(clinical_trial_data_to_load)
```

```python
# Merge the data into a single dataset
combined_data_df = pd.merge(tumor_data, drug_data, how='outer', on='Mouse ID')
```

```python
# Display the data table for preview
combined_data_df.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
Tumor_group_df = combined_data_df.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)']
avg_tumor_group_df = Tumor_group_df.mean()
avg_tumor_group_df.head()
```

```python
# Convert to DataFrame
Tumor_vol_df = pd.DataFrame(avg_tumor_group_df)
# Preview Data Frame
Tumor_vol_df.head(6)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_error = Tumor_group_df.sem()
```

```python
# Convert to DataFrame 
tumor_error_df = pd.DataFrame(tumor_error)
```

```python
# Preview DataFrame 
tumor_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
Tumor_vol_df = Tumor_vol_df.reset_index()

pivot_Tumor_vol_df = Tumor_vol_df.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')

# Preview that Reformatting worked
pivot_Tumor_vol_df.head()
```

```python
# Generate the Plot (with Error Bars)
drug_format_list = [('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black'),('Propriva','h','magenta'),('Ramicane','X','yellow'),('Stelasyn','P','chartreuse'),('Zoniferol','8','peru')]
for drug,marker,colors in drug_format_list:
    ste = tumor_error[drug]
    tumor_treatment_plt = plt.errorbar(pivot_Tumor_vol_df.index,pivot_Tumor_vol_df[drug],ste,
                                       fmt=marker,ls='--',color=colors,linewidth=0.5)
plt.legend(loc='upper left')
plt.title('Tumor Size Response to Treatment')
plt.xlabel('Days of Treatment')
plt.ylabel('Tumor Volume (mm3)')
plt.grid(color="0.75", linestyle="-") 
plt.rcParams['figure.figsize'] = (15,10)

# Save the Figure
plt.savefig('reponse_by_treatment.jpg')
```

```python
# Show the Figure
plt.show()
```

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
meta = combined_data_df.groupby(['Drug','Timepoint'])['Metastatic Sites']
avg_meta_df = meta.mean()
# Convert to DataFrame
meta_response_df = pd.DataFrame(avg_meta_df) 
# Preview DataFrame
meta_response_df.head()
```

```python
# Store the Standard Error  associated with Met. Sites Grouped by Drug and Timepoint 
ste_meta =meta.sem()

# Convert to DataFrame
ste_meta_df = pd.DataFrame(ste_meta)
# Preview DataFrame
ste_meta_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
meta_response_df = meta_response_df.reset_index()
pivot_metastatic_sem_df = meta_response_df.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites')
# Preview that Reformatting worked
pivot_metastatic_sem_df.head()
```

```python
# Generate the Plot (with Error Bars)
drug_format_list = [('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black'),('Propriva','h','magenta'),('Ramicane','X','yellow'),('Stelasyn','P','chartreuse'),('Zoniferol','8','peru')]
for drug,marker,colors in drug_format_list:
    ste = ste_meta[drug]
    met_treatmeant_plt = plt.errorbar(pivot_metastatic_sem_df.index,pivot_metastatic_sem_df[drug],ste,fmt=marker,ls='--',color=colors,linewidth=0.5,)
    
plt.legend(loc='upper left')
plt.title('MetaStatic Response to Treatment')
plt.xlabel('Days of Treatment')
plt.ylabel('Tumor Volume (mm3)')
plt.grid(color="0.75", linestyle="-")     
plt.rcParams['figure.figsize'] = (15,10)
# Save the Figure
plt.savefig('metastatic_response.jpg')

# Show the Figure
plt.show()
```

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_grouped_df = combined_data_df.groupby(['Drug','Timepoint'])['Mouse ID']
count_mouse_df = mouse_grouped_df.count() 

# Convert to DataFrame
count_mouse_df = pd.DataFrame(count_mouse_df)
count_mouse_df=count_mouse_df.rename(columns={'Mouse ID':'Mouse Count'})

# Preview DataFrame
count_mouse_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
count_mouse_df = count_mouse_df.reset_index()
pivot_count_mouse_df = count_mouse_df.pivot(index='Timepoint',columns='Drug',values='Mouse Count')

# Preview the Data Frame
pivot_count_mouse_df.head()
```

```python
# Generate the Plot (Accounting for percentages)
for drug,marker,colors in drug_format_list:
    total_mouse = pivot_count_mouse_df[drug][0]
    survival_rate = (pivot_count_mouse_df[drug]/total_mouse)*100
    survival_treatment_plt = plt.plot(pivot_count_mouse_df.index,survival_rate,
                                       marker=marker,ls='--',color=colors,linewidth=0.5)

# Show the Figure
plt.legend(loc='bottom left')
plt.title('Survival During Treatment')
plt.xlabel('Times (Days)')
plt.ylabel('Survival Rate (%)')
plt.grid()
plt.rcParams['figure.figsize'] = (15,10)
# Save the Figure
plt.savefig("Survival_rates.jpg")
# Show the Figure
plt.show()
```

```python
# Calculate the percent changes for each drug
percentage_change = (pivot_Tumor_vol_df.iloc[-1])-(pivot_Tumor_vol_df.iloc[0])

# Display the data to confirm
percentage_change
```

```python
# Store all Relevant Percent Changes into a Tuple
passing = percentage_change < 0

# set figure dimensions for easy reading
plt.figure(figsize = (15, 10))

# Orient widths. Add labels, tick marks, etc. 
drug_list = ['Capomulin','Infubinol','Ketapril','Placebo','Propriva','Ramicane','Stelasyn']
change_list = [(percentage_change[durg])for durg in drug_list]
change_plt = plt.bar(drug_list,change_list,width=-1,align='edge',color=passing.map({True:'g',False:'r'}))
plt.grid()
plt.ylim(-30,70)
plt.ylabel('% Tumor Volume Change')
plt.title('Tumor Change over 45 Day Treatment')

# Use functions to label the percentages of changes
def labelplot(rects):
    for rect in rects:
        height = rect.get_height()
        if height > 0:
            label_position = 2
        else:
            label_position = -8
        plt.text(rect.get_x() + rect.get_width()/2., label_position,
                '%d' % int(height)+'%',color='white',
                ha='center', va='bottom')
                
# Call functions to implement the function calls
labelplot(change_plt)

# Save the Figure
plt.savefig("Tumor_Change_45days.jpg")

# Show the Figure
plt.show()
```
